```python
import pandas as pd
from collections import Counter
import secret
```

## Importing and merging data

```python
df1 = pd.read_csv('data archive\\property_data.csv')
df2 = pd.read_csv('data archive\\property_data_2.csv')
df = pd.concat([df1,df2])
df = df.drop(df.columns[[0]], axis=1) 
df = df.reset_index(drop=True)
df
```

## Cleaning bds column

```python
bds = set(df['bds'])
map = {}
for i in bds:
    map[i] = i.split()[0]
map
```

```python
df['bds'] = df['bds'].map(map)
df
```

## Cleaning cost column

```python
df['cost'] = df['cost'].apply(lambda x: int(x.strip(',$/mo+').replace(',', '')))
df
```
